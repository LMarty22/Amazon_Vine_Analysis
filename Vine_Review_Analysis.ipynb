```python
#Dependencies
import pandas as pd
import os
```

```python
#Load in CSV

vines_table_df = pd.read_csv("vine_table.csv")

vines_table_df
```

```python
vines_table_df.count()
```

```python
# rows with 20 revirews or greater
vines_twenty_reviews_df = vines_table_df[vines_table_df["total_votes"] >=20]
vines_twenty_reviews_df
```

```python
#2.Filter new df, to retrieve all rows where the number of helpful votes 
#divided by the total votes is equal to or greater than 50%
vine_total_vote = vines_twenty_reviews_df["total_votes"]
#vine_total_vote.head()
vine_helpful_votes = vines_twenty_reviews_df["helpful_votes"]
#vine_helpful_votes
helpful_votes = vines_twenty_reviews_df[(vine_helpful_votes/vine_total_vote) >=.50]
helpful_votes
```

```python
#3. Retrieve all the rows where a review was wrtiten as part of the Vine program

vine_program_yes = helpful_votes[helpful_votes["vine"] == "Y"]
vine_program_yes
```

```python
#4. Retrieve all the rows where a review was wrtiten as part of the Vine program

vine_program_no = helpful_votes[helpful_votes["vine"] == "N"]
vine_program_no
```

```python
#5.
paid_reviews = len(vine_program_yes)
paid_reviews
```

```python
#5.
unpaid_reviews = len(vine_program_no)
unpaid_reviews
```

```python
paid_five_star_reviews = len(vine_program_yes[vine_program_yes["star_rating"]==5])
paid_five_star_reviews
```

```python
unpaid_five_star_reviews = len(vine_program_no[vine_program_no["star_rating"]==5])
unpaid_five_star_reviews
```

```python
perc_five_star_paid = paid_five_star_reviews/paid_reviews * 100
perc_five_star_paid
```

```python
perc_five_star_unpaid = unpaid_five_star_reviews/unpaid_reviews * 100
perc_five_star_unpaid
```
